# Turotial to obtain the PSF of a 30-m telescope using HyperCAT

This notebook shows how to obtain the PSF of a 30-m telescope using HyperCAT

```python
#HyperCAT
import sys
sys.path.append('../hypercat')
import psf_real  #module to obtain thePSF of a 30-m telescope

#3rd party
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Let's obtain the PSF of the TMT at 2.2 microns

```python
telescope = 'TMT'
wavelength = 2.2 # in microns

pxscale , psf = psf_real.psf_real(telescope,wavelength)

fig, ax1 = plt.subplots(1,1,figsize=(5,5))

fov = psf.shape[0]*pxscale/2./1000.  #FOV of the image

ax1.imshow(np.log10(psf),origin='lower',interpolation='Nearest', extent=[-fov,fov,-fov,fov])
ax1.set_title(np.str(wavelength)+' $\\mu$m')
ax1.set_xlabel('Offset (")')
ax1.set_ylabel('Offset (")')
```

Let's obtain the PSF of the E-ELT at 2.2 microns

```python
telescope = 'EELT'
wavelength = 2.2 # in microns

pxscale , psf = psf_real.psf_real(telescope,wavelength)


fig, ax1 = plt.subplots(1,1,figsize=(5,5))

fov = psf.shape[0]*pxscale/2./1000.  #FOV of the image

ax1.imshow(np.log10(psf),origin='lower',interpolation='Nearest', extent=[-fov,fov,-fov,fov])
ax1.set_title(np.str(wavelength)+' $\\mu$m')
ax1.set_xlabel('Offset (")')
ax1.set_ylabel('Offset (")')
```
